```python
def factorial(n):
     if n == 0:
        return 1
     elif n < 0:
        return "Factorial is non definend"

     else:
         return n * factorial(n - 1)


  # number = 5
  # result = factorial(number)
number = int(input ("Enter the number: "))
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def check_odd_even(number):
   if number % 2 == 0:
    return "Even"
   else:
    return "odd"


num1 = 5
num2 = 10
num3 = 7

print(f"The number {num1} is {check_odd_even (num1)}. ")
print(f"the number {num2} is {check_odd_even (num2)}. ")
print(f"the number {num3} is {check_odd_even (num3)}. ")
```

```python
def is_palindrome(number):
  num_str = str(number)
  reversed_str = num_str[::-1]
  if num_str == reversed_str:
    return True
  else:
    return False

try:
  user_input = int(input("enter a number: "))

  if is_palindrome(user_input):
    print(f"{user_input} is palindrome.")
  else:
    print(f"{user_input} is not a palindrome.")
except ValueError:
   print("Invalid input. please enter an integer.")
```
